```python
#3,categorial estimate plot
#1.barplot
```

```python
ax=sns.barplot(df,x='sex',y='total_bill',hue='smoker',estimator=np.min,errorbar=None)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom',color='red')
```

```python
female_smoker=df[(df['sex']=='Female') & (df['smoker']=='Yes')]
```

```python
min_female=female_smoker['total_bill'].np.std()
```

```python
#plotting pairwise relation
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.pairplot(iris,hue='species')
```

```python
sns.PairGrid(iris).map_diag(sns.barplot).map_offdiag(sns.lineplot)
```

```python
g=sns.PairGrid(iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
g=sns.PairGrid(iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```
